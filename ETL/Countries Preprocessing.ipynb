```python
# Import dependencies
import pandas as pd
```

```python
# Load in the countries.csv file
datafile = "CSV_In_Progress/countries.csv"

# Read the CSV into a DataFrame
countries_df = pd.read_csv(datafile)

# Test script to ensure above completed correctly
countries_df
```

```python
# Drop all rows that contain NAs
countries_nona_df = countries_df.dropna()

# Test script to ensure above completed correctly
countries_nona_df
```

```python
# Drop unneeded data in columns Importance and Altitude
f_countries_df = countries_nona_df.drop(['Importance', 'Altitude'], axis = 1)

# Test script to ensure above completed correctly
f_countries_df
```

```python
# Export preprocessed data to new CSV
output_path = "CSV_In_Progress/countries_preprocessed.csv"
f_countries_df.to_csv(output_path, index = False)
```

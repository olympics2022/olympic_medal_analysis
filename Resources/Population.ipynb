```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Read Csv. Skip first 3 rows to avoid parser error
pop_file = pd.read_csv('Resources/API_SP.POP.TOTL_DS2_en_csv_v2_3469297.csv')
pop_file
```

```python
# Drop NaN values
pop_file.dropna(inplace=True)
```

```python
# Drop columns not needed and use melt() funt to keep cols need & name them
temp = pop_file.drop(columns=['Indicator Name', 'Indicator Code'])
clean_pop =temp.melt(id_vars=['Country Code', 'Country Name'], var_name = 'Year', value_name = 'Population') 
clean_pop
```

```python
# Sum of NA values
clean_pop.isna().sum()
```

```python
# Export clean Population df as a CSV
clean_pop.to_csv('Country_population.csv', index=False)
```

```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the dataset
olympic_df = pd.read_csv('../CSV_In_Progress/Machine_Learning_dataset.csv')
olympic_df = olympic_df.drop(['Unnamed: 0'], axis=1)
```

```python
# Sort for Winter Olympics Results
winter_olympic_df = olympic_df[olympic_df['Season'] == 'Winter']
winter_olympic_df
```

```python
# Drop unused columns
X = winter_olympic_df.drop(['NOC', 'Year', 'Season', 'Gold', 'Silver', 'Bronze', 'country_medal_total', 'olympic_total', 'medal_share'], axis=1)
```

```python
X.shape
```

```python
y = winter_olympic_df.country_medal_total
```

```python
# Create a model with scikit-learn
model = LinearRegression()
```

```python
# Fit the data into the model
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
# Import Dependencies
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
from statsmodels.graphics.regressionplots import abline_plot
from sklearn.linear_model import LinearRegression
```

```python
olympic_df = pd.read_csv('../Resources/Machine_Learning_datasetv3.csv')
summer2020_olympic_df = olympic_df[olympic_df['Season'] == 'Summer']
summer2020_olympic_df['lnpopulation'] = np.log2(summer2020_olympic_df['Population'])
summer2020_olympic_df = summer2020_olympic_df.rename(columns={'GDP per capita': 'gdpperc'})
summer2020_olympic_df
```

```python
# Multiple Linear Regression model
model_mlr = sm.ols(formula='medal_share ~ gdpperc + athlete_share + Host + lnpopulation', data=summer2020_olympic_df).fit()
model_mlr.summary()
```
